```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

## Loading DataFrame

```python
df=pd.read_csv('honeyproduction 1998-2021.csv')
df.head()
```

## Converted big values in readable format.

```python
df['totalprod']=df['totalprod']/1000000   ## converted into millions of pounds
df['prodvalue']=df['prodvalue']/1000000   ## converted into millions of dollars
df['numcol']=df['numcol']/1000            ## converted into thousand of colony
df['stocks']=df['stocks']/1000000         ## converted inot millions of pounds
df.head()
```

## 1.How has honey production yield changed from 1998 to 2021?

### * First hoeny production decrease from 1998 to 2003 then increases from 2004 to 2006. then again townward till 2021.
### * In 2006 maximum honey production was done. After 2006 great fall in produciton.
### * In 2009 minimum honey productiom was done.

### * As shown below in chart

```python
df1=df.groupby('year')['totalprod'].median()
plt.figure(figsize=(10,6))
df1.plot.line(marker='*',color='maroon')
plt.xlabel('Years (1998-2021)',size=12,color='green')
plt.ylabel('Average Total Production Per Year (In millions of pounds)',size=12,color='green')
plt.title('Year Wise Average Total Production',size=15,color='green')
plt.xticks(rotation=90)
plt.show()
```

## 2.Overtime, what are the major production trends across the states?

###  * Across all states production of honey is decreasing over time.
###  * States California, North Dakota, South Dakota, florida are Top 5 honey producing States in USA.

```python
df2=df[['State','totalprod','year']]
plt.figure(figsize=(8,5))
gra=sns.FacetGrid(data=df2,col='State')
gra.map(sns.lineplot,'year','totalprod')
plt.show()
```

# 3.Does the data show any trends in terms of the number of honey producing colonies and yield per colony before 2006,which was when concern over Colony Collapse Disorder spread nationwide?

### * Yield per colony depend upon then no. of colonies.
### * if no. of colony is more then yield per colony is also more.

```python
df3=df[['numcol','yieldpercol','year']][df['year']<2006].groupby(['year'])[['yieldpercol','numcol']].median()
sns.set(rc={'figure.figsize':(8,5)})
df3.plot.bar()
sns.set_style('whitegrid')
sns.set_context('paper')
plt.xticks(rotation=90)
plt.title('Yield Per Colony(In pounds) Vs No. of Honey Producing Colony(In Thousands) before 2006',size=15)
plt.legend(loc='right',bbox_to_anchor=(0.2,0.4,1,1))
plt.show()
```

## 4.Are there any patterns that can be observed between total honey production and value of production every year?

### * Value of the production and Total production are opposite when one increases other decreases.
### * Over the time Value of production is  increasing.
### * Over the time Total Production is decreasing.

```python
df4=df[['totalprod','prodvalue','year']].groupby('year')[['totalprod','prodvalue']].median()
plt.figure(figsize=(10,8))
df4.plot.line()
plt.xticks(rotation=90)
plt.title('Total Production(In millions Of Poounds) Vs Value of Production(In millions Of Dollars) Over Time)',size=15,color='green')
plt.ylabel('Value in Millions',size=12,color='green')
plt.show()
```

## 5.How has the value of production, which in some sense could be tied to demand, changed every year?

### * As shown in chart below the Production Value is continuously increasing over time.
### * So demand is tied by increasing the value of production.


```python
df['demand']=df['totalprod']-df['stocks']        ## created demand column in millions of pound
df5=df[['prodvalue','year','demand']].groupby('year')[['prodvalue','demand']].median()
df5.plot.bar()
plt.title('Production Value(In Millions of dollars), Demand(In Millions of pounds) over time',size=15,color='green')
plt.ylabel('Value in Millions',size=12,color='green')
plt.show()
```

## 6.Constructs the related plots using Seaborn and Matplot apply customization and derive insights from the visualization.

### * Top 10 honey producing states
### * North Dakota is No.1

```python
df6=df.groupby('State')['totalprod'].median().sort_values(ascending=False).head(10)
df6.plot.bar()
plt.title("Top 10 Honey Producing states",size=15)
plt.ylabel('Total Producion in Millions of pounds',size=12)
plt.show()
```

### * Top 10 Honey producion year.
### * 2006 is top production year for honey

```python
df7=df.groupby('year')['totalprod'].median().sort_values(ascending=False).head(10)
df7.plot.bar()
plt.title('Top 10 Honey production year',size=15)
plt.ylabel('Total Production in Millions of pounds',size=12)
plt.show()
```

### * Top 10 honey producing colony

```python
df8=df.groupby('State')[['numcol']].median().sort_values(by='numcol',ascending=False).head(10)
df8.plot.bar()
plt.ylabel('No of colony in thousands',size=12)
plt.title('Top 10 honey producing colony',size=15)
plt.show()
```
